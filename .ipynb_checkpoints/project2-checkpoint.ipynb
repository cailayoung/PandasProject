### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep= "\t")
chipo
```

### Step 4. How many products cost more than $10.00?

```python
#chipo.dtypes
nodollar= chipo["item_price"].str.replace("$","", regex=True)
chipo["New"] = nodollar.astype(float)
x= chipo[chipo["New"]>10]
x
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
itemprice = chipo[["item_name","item_price"]]
itemprice
```

### Step 6. Sort by the name of the item

```python
itemprice.sort_values(by="item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
quantity = chipo[["quantity", "New"]]
sortd= quantity.sort_values(by="New", ascending= False)
sortd.max()
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
quantity = chipo[["item_name", "quantity"]]
x= quantity[(quantity["item_name"]== "Veggie Salad Bowl")]
y= x.value_counts()
y
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
filt=(chipo["item_name"] == "Canned Soda") & (chipo["quantity"] > 1)
times = chipo.loc[filt].shape[0]
times
```
